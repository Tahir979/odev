```python
## modüller yüklendi
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import math

##.csv dosyaları yüklendi
nRowsRead = 1000
df15 = pd.read_csv('2015.csv', delimiter=',', nrows = nRowsRead)
df15.dataframeName = '2015.csv'

df16 = pd.read_csv('2016.csv', delimiter=',', nrows = nRowsRead)
df16.dataframeName = '2016.csv'

df17 = pd.read_csv('2017.csv', delimiter=',', nrows = nRowsRead)
df17.dataframeName = '2017.csv'
```

```python
# En yüksek mutluluk puanına sahip 10 ülke
top10 = df15.sort_values(by = 'Happiness Score', ascending = False).head(10)
plt.figure(figsize=(20, 10))
sns.barplot(x = 'Country', y = 'Happiness Score', data = top10)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Ekonomisi en iyi durumda olan 10 ülke
top10_gdp = df15.sort_values(by = 'Economy (GDP per Capita)', ascending = False).head(10)
plt.figure(figsize=(25, 10))
sns.barplot(x = 'Country', y = 'Economy (GDP per Capita)', data = top10_gdp)
plt.xticks(rotation = 90)
plt.show()
```

```python
# En uzun ömürlü insanlara sahip olması beklenen 10 ülke
top10_social = df15.sort_values(by = 'Health (Life Expectancy)', ascending = False).head(10)
plt.figure(figsize=(25, 10))
sns.barplot(x = 'Country', y = 'Health (Life Expectancy)', data = top10_social)
plt.xticks(rotation = 90)
plt.show()
```

```python
#En cömert 10 ülke
top10_health = df15.sort_values(by = 'Generosity', ascending = False).head(10)
plt.figure(figsize=(25, 10))
sns.barplot(x = 'Country', y = 'Generosity', data = top10_health)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Özgürlükler bakımından en önde 10 ülke
top10_freedom = df15.sort_values(by = 'Freedom', ascending = False).head(10)
plt.figure(figsize=(25, 10))
sns.barplot(x = 'Country', y = 'Freedom', data = top10_freedom)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Hükümete güven konusundaki ilk 10 ülke
top10_social = df15.sort_values(by = 'Trust (Government Corruption)', ascending = False).head(10)
plt.figure(figsize=(25, 10))
sns.barplot(x = 'Country', y = 'Trust (Government Corruption)' , data = top10_social)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Mutluluk puanlarının genel dağılımı
sc15=df15['Happiness Score'].mean()
sc16=df16['Happiness Score'].mean()
sc17=df17['Happiness.Score'].mean()
Hscore=np.array([sc15,sc16,sc17])
xx=np.arange(2015,2020,1)
yy=Hscore

plt.figure(figsize=(10,5))
plt.grid(True)
plt.title("Hapiness Score")
plt.xlabel("Year")
plt.ylabel("Hapiness")
plt.xticks(xx, xx)

plt.plot(xx,yy)
plt.show()
```

2015-2016 yılları arasındaki değişim

```python
import math
df16_arranged = df16.T.drop(["Lower Confidence Interval","Upper Confidence Interval"],inplace=False).T

df = abs(df15["Happiness Score"] -df16_arranged["Happiness Score"])

df = pd.DataFrame(df)
df_to_plot = pd.concat([df16["Country"], df],axis=1)
df_to_plot.dropna(inplace=True)

plt.figure(figsize=(25, 10))
sns.barplot(x='Country', y='Happiness Score', data=df_to_plot)
plt.xticks(rotation=90)
plt.show()
```

2016-2017 yılları arasındaki değişim

```python
import math
df16_arranged = df16.T.drop(["Region", "Lower Confidence Interval","Upper Confidence Interval"],inplace=False).T
df17_arranged = df17.T.drop(["Whisker.low","Whisker.high"],inplace=False).T

df16_arranged=df16.rename(columns={"Economy (GDP per Capita)":"Economy","Health (Life Expectancy)":"Health",
                      "Trust (Government Corruption)":"Trust"
                      })

df17_arranged=df17.rename(columns={"Happiness.Rank":"Happiness Rank", "Happiness.Score":"Happiness Score", "Economy..GDP.per.Capita.":"Economy","Health..Life.Expectancy.":"Health",
                      "Trust..Government.Corruption.":"Trust"
                      })

df = abs(df16_arranged["Happiness Score"] -df17_arranged["Happiness Score"])

df = pd.DataFrame(df)
df_to_plot = pd.concat([df17["Country"], df],axis=1)
df_to_plot.dropna(inplace=True)

plt.figure(figsize=(25, 10))
sns.barplot(x='Country', y='Happiness Score', data=df_to_plot)
plt.xticks(rotation=90)
                      
```
